```python
import matplotlib.pyplot as plt

# Datos de ejemplo
x = [5, 7, 8, 7, 6, 9, 5, 6, 7, 8]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

# Crear gráfico de dispersión
plt.scatter(x, y, color="blue", marker="o", s=100)  # s es el tamaño de los puntos

# Títulos y etiquetas
plt.title("Ejemplo de gráfico de dispersión")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")

# Mostrar gráfico
plt.show()
```
